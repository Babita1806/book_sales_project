```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data

df = pd.read_csv("D:\\dataset\\Books_Data_Clean.csv")
df
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df2 = df.copy()
df2.head(5)
```

```python
df2.dropna(subset= "Book Name", inplace= True)
```

```python
df2.duplicated().sum()
```

```python
df2.nunique()
```

```python
plt.hist(df2["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distrubution of Publishing Year")
plt.show
```

```python
df2["genre"].value_counts().plot(kind= "bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of books in each genre")
plt.show
```

```python
df2.groupby("Author")["Book_average_rating"].mean().sort_values(ascending= False)
```

```python
sns.boxplot(x = "genre", y = "Book_ratings_count", data = df2)
plt.xlabel("Genre")
plt.ylabel("Book Rating Count")
plt.title("Boxplot of Book Rating Count for Each Genre")
plt.show
```

```python
plt.scatter(df2["sale price"], df2["units sold"])
plt.xlabel("Sales Price")
plt.ylabel("Unit Sold")
plt.title("Sactter plot of Sales prive Vs Unit sold")
plt.show
```

```python
Language_counts =  df2["language_code"].value_counts()
```

```python
plt.pie(Language_counts, labels= Language_counts.index, startangle= 90, autopct= "%1.1f%%" )
plt.title("Language Distribution of Books")
plt.show
```

```python
df2.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending= False)
```

```python
total_gross_sales_by_author = df2.groupby("Author")["gross sales"].sum()
```

```python
total_gross_sales_by_author.sort_values(ascending= False).head(10).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total gross sales for each Author")
plt.show()
```
